```python
import pandas as pd
import requests
from PIL import Image
from io import BytesIO
import os
import pytesseract

# Path to your CSV file
csv_file_path = '/Users/tanmoyroy/Desktop/student_resource 3/dataset/two/test.csv'

# Read the CSV file into a DataFrame
df = pd.read_csv(csv_file_path)

# Directory to save images
base_dir = '/Users/tanmoyroy/Desktop/student_resource 3/dataset/two'
output_dir = os.path.join(base_dir, 'images_content')
os.makedirs(output_dir, exist_ok=True)

# If you're on Windows, specify the path to the Tesseract executable
# For Mac or Linux, ensure tesseract is in your PATH
# Uncomment and set the path if needed
# pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'

# Function to download and save image
def download_image(url, filename):
    try:
        response = requests.get(url)
        response.raise_for_status()  # Check if the request was successful
        image = Image.open(BytesIO(response.content))
        image.save(filename)
        print(f"Image saved: {filename}")
        return image
    except Exception as e:
        print(f"Failed to download {url}. Error: {e}")
        return None

# Function to extract text from an image
def extract_text_from_image(image):
    try:
        text = pytesseract.image_to_string(image)
        return text.strip()  # Remove leading and trailing whitespace
    except Exception as e:
        print(f"Failed to extract text from image. Error: {e}")
        return ""

# Select the top 100 rows
df_top_100 = df.head(100)

# List to store results
results = []

# Iterate over the top 100 rows of the DataFrame, download images, and extract text
for index, row in df_top_100.iterrows():
    image_url = row['image_link']
    group_id = row['group_id']
    entity_name = row['entity_name']
    
    # Generate a filename based on index or any unique identifier
    filename = os.path.join(output_dir, f'image_{index}.jpg')
    
    # Download and save the image
    image = download_image(image_url, filename)
    
    if image:
        # Extract text from the downloaded image
        extracted_text = extract_text_from_image(image)
        
        # Append results to the list
        results.append({
            'index': index,
            'group_id': group_id,
            'entity_name': entity_name,
            'image_content': extracted_text
        })

# Create a DataFrame with the results
results_df = pd.DataFrame(results)

# Save the DataFrame to a new CSV file
results_csv_path = os.path.join(base_dir, 'image_text_results.csv')
results_df.to_csv(results_csv_path, index=False)

print(f"Results saved to: {results_csv_path}")
```

```python
import pandas as pd

# Path to the original CSV file
input_csv_path = '/Users/tanmoyroy/Desktop/student_resource 3/dataset/two/image_text_results.csv'

# Path to save the cleaned CSV file
output_csv_path = '/Users/tanmoyroy/Desktop/student_resource 3/dataset/two/cleaned_sample_test.csv'

# Read the CSV file into a DataFrame
df = pd.read_csv(input_csv_path)

# Function to clean the image_content column
def clean_text(text):
    if pd.isna(text):
        return text
    text = text.replace('\n', ' ')  # Replace newlines with spaces
    text = text.replace('\\', ' ')  # Replace backslashes with spaces
    return text.strip()  # Remove leading and trailing whitespace

# Apply the cleaning function to the image_content column
df['image_content'] = df['image_content'].apply(clean_text)

# Save the cleaned DataFrame to a new CSV file
df.to_csv(output_csv_path, index=False)

print(f"Cleaned CSV saved to: {output_csv_path}")
```
